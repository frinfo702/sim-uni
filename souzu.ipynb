```python
# 
def plot_vector_field(dx: dx, dy:dy):

    # 
    x = np.linspace(-2, 2, 20)
    y = np.linspace(-2, 2, 20)
    X, Y = np.meshgrid(x, y)

    dR = dx(X, Y)
    dJ = dy(X, Y)

    plt.quiver(X, Y, dR, dJ, scale=25)
    plt.xlabel('R')
    plt.ylabel('J')
    plt.title(f"vector field")
    plt.grid()
    plt.show()
```
