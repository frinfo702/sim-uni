```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-8, 8, 200)
y = x**2/4
```

```python
plt.figure(figsize=(10,6))

# 
plt.plot(x, y, color='black', linewidth=1)

# a-1
plt.fill_between(x, y, where=(x>0)&(y>0), color='red', alpha=0.3, label="^2>4 and >0 and >0")

# a-2
plt.fill_between(x, y, where=(x<0)&(y>0), color='orange', alpha=0.3, label="^2>4 and <0 and >0")

# a-3
plt.fill_between(x, 0, -8, color='yellow', alpha=0.3, label="^2>4 and =<0")

# b-1
plt.vlines(x=0, ymin=0, ymax=plt.ylim()[1], color="green", linewidth=2, label="^2<4 and =0")

# b-2
plt.fill_between(x, y, plt.ylim()[1], where=(x>0), color='blue', alpha=0.3, label="^2>4 and >0")

# b-3
plt.fill_between(x, y, plt.ylim()[1], where=(x<0), color='purple', alpha=0.3, label="^2>4 and <0")

plt.xlabel('')
plt.ylabel('')
plt.xlim((-8, 8))
plt.ylim((-8, 8))
plt.title('- field')
plt.legend()
plt.savefig("-field.png")
plt.show()
```
