```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-8, 8, 200)
y = x**2/4
```

```python
plt.figure(figsize=(10,6))

# 全体の曲線
plt.plot(x, y, color='black', linewidth=1)

# 領域a-1
plt.fill_between(x, y, where=(x>0)&(y>0), color='red', alpha=0.3, label="τ^2>4∆ and τ>0 and ∆>0")

# 領域a-2
plt.fill_between(x, y, where=(x<0)&(y>0), color='orange', alpha=0.3, label="τ^2>4∆ and τ<0 and ∆>0")

# 領域a-3
plt.fill_between(x, 0, -8, color='yellow', alpha=0.3, label="τ^2>4∆ and Δ=<0")

# 領域b-1
plt.vlines(x=0, ymin=0, ymax=plt.ylim()[1], color="green", linewidth=2, label="τ^2<4∆ and τ=0")

# 領域b-2
plt.fill_between(x, y, plt.ylim()[1], where=(x>0), color='blue', alpha=0.3, label="τ^2>4∆ and τ>0")

# 領域b-3
plt.fill_between(x, y, plt.ylim()[1], where=(x<0), color='purple', alpha=0.3, label="τ^2>4∆ and τ<0")

plt.xlabel('τ')
plt.ylabel('Δ')
plt.xlim((-8, 8))
plt.ylim((-8, 8))
plt.title('τ-Δ field')
plt.legend()
plt.savefig("τ-Δfield.png")
plt.show()
```
