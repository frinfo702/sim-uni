```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def euler_method(a, b, c, d, R0, J0, T, dt=0.01):
    """
    オイラー法を用いて微分方程式を数値的に解く関数
    dR/dt = aR + bJ
    dJ/dt = cR + dJ
    """
    num_steps = int(T / dt) + 1
    t = np.linspace(0, T, num_steps)
    R = np.zeros(num_steps)
    J = np.zeros(num_steps)
    
    R[0] = R0
    J[0] = J0
    
    for i in range(1, num_steps):
        dR_dt = a * R[i-1] + b * J[i-1]
        dJ_dt = c * R[i-1] + d * J[i-1]
        
        R[i] = R[i-1] + dR_dt * dt
        J[i] = J[i-1] + dJ_dt * dt
    
    return t, R, J

def plot_results(t, R, J):
    """
    シミュレーション結果をグラフ表示する関数
    """
    # 時間tに対するR(t)とJ(t)のグラフ
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.plot(t, R, label='R(t)')
    plt.plot(t, J, label='J(t)')
    plt.xlabel('時間 t')
    plt.ylabel('値')
    plt.title('R(t) と J(t) の時間変化')
    plt.legend()
    
    # RとJの関係図
    plt.subplot(1, 2, 2)
    plt.plot(R, J)
    plt.xlabel('R')
    plt.ylabel('J')
    plt.title('状態ベクトル (R, J) の軌道')
    
    plt.tight_layout()
    plt.show()

    
```

```python
# パラメータの設定
a, d = 0, 0
b, c = 0.5, 0.5
R0, J0 = 1.0, 1.0
T = 20.0
dt = 0.01

t, R, J = euler_method(a, b, c, d, R0, J0, T, dt)
plot_results(t, R, J)
```
