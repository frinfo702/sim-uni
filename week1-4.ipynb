```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 使用する関数群
def simulate_differential_equations(a, b, c, d, R0, J0, T, dt=0.01):
    """
    オイラー法を用いて恋愛微分方程式を数値的に解く関数

    パラメータ:
    a, b, c, d: システムパラメータ
    R0, J0: 初期値
    T: シミュレーション時間
    dt: 時間刻み幅
    """
    num_steps = int(T / dt) + 1
    t = np.linspace(0, T, num_steps)
    R = np.zeros(num_steps)
    J = np.zeros(num_steps)
    
    R[0] = R0
    J[0] = J0
    
    for i in range(1, num_steps):
        dR_dt = a * R[i-1] + b * J[i-1]
        dJ_dt = c * R[i-1] + d * J[i-1]
        
        R[i] = R[i-1] + dR_dt * dt
        J[i] = J[i-1] + dJ_dt * dt
    
    data = pd.DataFrame({'t': t, 'R': R, 'J': J})
    return data

def plot_results(data):
    """
    シミュレーション結果をグラフ表示する関数

    パラメータ:
    data: シミュレーション結果を含むDataFrame
    """
    plt.figure(figsize=(12, 5))

    # 時間tに対するR(t)とJ(t)のグラフ
    plt.subplot(1, 2, 1)
    plt.plot(data['t'], data['R'], label='R(t)')
    plt.plot(data['t'], data['J'], label='J(t)')
    plt.xlabel('時間 t')
    plt.ylabel('値')
    plt.title('時間に対するR(t)とJ(t)')
    plt.legend()

    # RとJの相関グラフ
    plt.subplot(1, 2, 2)
    plt.plot(data['R'], data['J'])
    plt.xlabel('R(t)')
    plt.ylabel('J(t)')
    plt.title('状態ベクトルの軌道 (R vs J)')

    plt.tight_layout()
    plt.show()
```

```python
# パラメータの設定
a = 1.0  
b = 0.5  
c = -0.5
d = 1.0  
R0 = 10.0 # 初期値 R(0)
J0 = 5.0  # 初期値 J(0)
T = 20.0  # シミュレーション時間

# シミュレーションの実行
simulation_data = simulate_differential_equations(a, b, c, d, R0, J0, T)

# 結果のプロット
plot_results(simulation_data)
```
